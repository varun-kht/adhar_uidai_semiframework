```python
import pandas as pd
enrolment_df = pd.read_csv(r"C:\Users\varun\Desktop\udai_adhar _framework\data\raw\enrolment.csv")
biometric_df = pd.read_csv(r"C:\Users\varun\Desktop\udai_adhar _framework\data\raw\biometric.csv")
demographic_df = pd.read_csv(r"C:\Users\varun\Desktop\udai_adhar _framework\data\raw\demographic.csv")
```

Normalization of columns

```python
def normalize_columns(df):
    df = df.copy()
    df.columns = (
        df.columns
        .str.strip()
        .str.lower()
        .str.replace(" ", "_")
    )
    return df

enrolment_df = normalize_columns(enrolment_df)
biometric_df = normalize_columns(biometric_df)
demographic_df = normalize_columns(demographic_df)
```

Mapping for common ground

```python
enrolment_df["month"] = pd.to_datetime(
    enrolment_df["date"],
    format="%d-%m-%Y",
    errors="coerce"
).dt.to_period("M").astype(str)

biometric_df["month"] = pd.to_datetime(
    biometric_df["date"],
    format="%d-%m-%Y",
    errors="coerce"
).dt.to_period("M").astype(str)

demographic_df["month"] = pd.to_datetime(
    biometric_df["date"],
    format="%d-%m-%Y",
    errors="coerce"
).dt.to_period("M").astype(str)
```

```python
enrolment_df = enrolment_df.drop(columns=["date"])
biometric_df = biometric_df.drop(columns=["date"])
demographic_df = demographic_df.drop(columns=["date"])
```

Aggregate to state and month

```python
enrolment_monthly = (
    enrolment_df
    .groupby(["state", "month"], as_index=False)
    .sum()
)

biometric_monthly = (
    biometric_df
    .groupby(["state", "month"], as_index=False)
    .sum()
)
demographic_monthly = (
    demographic_df
    .groupby(["state", "month"], as_index=False)
    .sum()
)
```

```python
final_df = enrolment_monthly.merge(
    biometric_monthly,
    on=["state", "month"],
    how="inner"
)

f_df = final_df.merge(
    demographic_monthly,
    on=["state", "month"],
    how="inner"
)

```
